```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/Churn_Modelling.csv')
```

```python
df.head()
```

```python
df.drop(columns = ['RowNumber','CustomerId','Surname'],inplace = True)
```

```python
df.head()
```

```python
df['Geography'].value_counts()
```

```python
df['Gender'].value_counts()
```

```python
df = pd.get_dummies(df,columns = ['Geography','Gender'],drop_first = True)
```

```python
df.head()
```

```python
x= df.drop(columns = ['Exited'])
y = df['Exited']

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(11,activation = 'sigmoid',input_dim = 11))
model.add(Dense(11,activation = 'sigmoid'))
model.add(Dense(1,activation = 'sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss = 'binary_crossentropy',optimizer = 'Adam',metrics = ['accuracy'])
```

```python
history = model.fit(x_train,y_train,epochs = 100,validation_split = 0.2, batch_size=50)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
y_pred_test = []
for i in y_pred:
    if i<=0.5:
        y_pred_test.append(0)
    else:
        y_pred_test.append(1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred_test)
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train','validation'],loc = 'upper left')
plt.show()
```

```python
# prompt: do the same with diabetes.csv as we do earlier with churn_modelling.csv and i want accuracy above 0.85

import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/diabetes.csv')
df.head()

x = df.drop(columns=['Outcome'])
y = df['Outcome']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

# Try a slightly more complex model to potentially improve accuracy
model = Sequential()
model.add(Dense(12, activation='relu', input_dim=x_train.shape[1])) # Using relu activation and dynamic input dim
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.summary()

model.compile(loss='binary_crossentropy', optimizer='Adam', metrics=['accuracy'])

# Increase epochs and potentially batch size to give the model more training
history = model.fit(x_train, y_train, epochs=200, validation_split=0.2, batch_size=32)

y_pred = model.predict(x_test)
y_pred_test = [1 if i > 0.5 else 0 for i in y_pred]

accuracy = accuracy_score(y_test, y_pred_test)
print(f"Accuracy: {accuracy}")

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```

# **Day-2  mnist classification**

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Flatten
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
x_test.shape
```

```python
y_train
```

```python
x_train.shape
```

```python
import matplotlib.pyplot as plt
plt.imshow(x_train[2])
```

```python
x_train, y_train
```

```python
x_train[0]
```

```python
x_train = x_train/255
x_test = x_test/255
```

```python
x_train[0]
```

```python
model = Sequential()

model.add(Flatten(input_shape = (28,28)))

model.add(Dense(128,activation = 'relu'))
model.add(Dense(32,activation = 'relu'))
model.add(Dense(10,activation = 'softmax'))
```

```python
model.summary()
```

```python
model.compile(loss = 'sparse_categorical_crossentropy', optimizer = 'Adam', metrics = ['accuracy'])
```

```python
history = model.fit(x_train,y_train,epochs = 25,validation_split = 0.2, verbose=1)
```

```python
y_prob = model.predict(x_test)
```

```python
y_pred = y_prob.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```
